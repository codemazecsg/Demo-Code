```
df = spark.read.json("dbfs:/FileStore/tables/Portfolios_20170102.json")
```

```
df.printSchema()
```

```
from pyspark.sql import functions as F
df2 = df.select(F.explode("Portfolios").alias("ports"))
```

```
display(df2)
```

```
df2.printSchema()
```

```
df3 = df2.select("ports.CustomerID", "ports.CustomerName", "ports.PortfolioCreateDate", "ports.PortfolioID", "ports.TaxID", F.explode("ports.Holdings").alias("Holdings")).withColumnRenamed("ports.CustomerID", "CustomerID").withColumnRenamed("ports.CustomerName", "CustomerName").withColumnRenamed("ports.PortfolioCreateDate", "PortfolioCreateDate").withColumnRenamed("ports.PortfolioID", "PortfolioID").withColumnRenamed("ports.TaxID", "TaxID")
```

```
df3.printSchema()
```

```
display(df3)
```

```
df4 = df3.select("CustomerID", "CustomerName", "PortfolioCreateDate", "PortfolioID", "TaxID", "Holdings.HoldingID", "Holdings.PurchaseDate", "Holdings.Quantity", "Holdings.Tickersymbol")
```

```
df4.printSchema()
```
